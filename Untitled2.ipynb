## Model 1: Job Demand Forecast, setup.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error
from sklearn.preprocessing import OrdinalEncoder
from lightgbm import LGBMRegressor

df = pd.read_csv("refined_postings.csv")

df = df.sample(n=5000, random_state=42)

df_model = df.dropna(subset=['normalized_salary', 'industry_name', 'formatted_work_type', 'state', 'location'])

upper_clip = df_model['normalized_salary'].quantile(0.99)
df_model = df_model[df_model['normalized_salary'] <= upper_clip]

features = ['industry_name', 'formatted_work_type', 'state', 'location']
target = 'normalized_salary'
X = df_model[features]
y = np.log1p(df_model[target])  

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_encoded = encoder.fit_transform(X)

X_indices = df_model.index
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_encoded, y, X_indices, test_size=0.2, random_state=42)

model = LGBMRegressor(n_estimators=100, max_depth=8, random_state=42)
model.fit(X_train, y_train)

y_pred_log = model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_test_real = np.expm1(y_test)

rmse = root_mean_squared_error(y_test_real, y_pred)
mae = mean_absolute_error(y_test_real, y_pred)
r2 = r2_score(y_test_real, y_pred)

print("✅ Results")
print("RMSE:", round(rmse, 2))
print("MAE:", round(mae, 2))
print("R² Score:", round(r2, 4))
```

## Visualization of Model 1.

## Ablation study of Model 1.

```python
feature_sets = {
    "All Features": ['industry_name', 'formatted_work_type', 'state', 'location'],
    "No industry_name": ['formatted_work_type', 'state', 'location'],
    "No formatted_work_type": ['industry_name', 'state', 'location'],
    "No state": ['industry_name', 'formatted_work_type', 'location'],
    "No location": ['industry_name', 'formatted_work_type', 'state']
}

ablation_results = []

for label, features in feature_sets.items():
    X = df_model[features]
    y = np.log1p(df_model['normalized_salary'])

    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    X_encoded = encoder.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

    model = LGBMRegressor(n_estimators=100, max_depth=8, random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_pred = np.expm1(y_pred)
    y_test_real = np.expm1(y_test)

    rmse = root_mean_squared_error(y_test_real, y_pred)
    mae = mean_absolute_error(y_test_real, y_pred)
    r2 = r2_score(y_test_real, y_pred)

    ablation_results.append({
        "Feature Set": label,
        "RMSE": round(rmse, 2),
        "MAE": round(mae, 2),
        "R²": round(r2, 4)
    })

ablation_df = pd.DataFrame(ablation_results)
print("🔍 Ablation Study Results")
display(ablation_df)
```

## Model 2: Skill Extraction, setup.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer, OrdinalEncoder
from sklearn.multioutput import MultiOutputClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report
import re
from collections import Counter

df_postings = pd.read_csv("refined_postings.csv")
df_skills = pd.read_excel("Skills.xlsx")

skill_set = set(df_skills['Element Name'].dropna().str.lower().unique())

def extract_skills(text):
    if pd.isna(text): return []
    text_clean = re.sub(r'[^a-zA-Z ]', ' ', text).lower()
    words = set(text_clean.split())
    return list(skill_set.intersection(words))

df_postings['all_text'] = df_postings['description_x'].fillna('') + ' ' + df_postings['skills_desc'].fillna('')
df_postings['matched_skills'] = df_postings['all_text'].apply(extract_skills)

N = 10
all_skills_flat = [skill for sublist in df_postings['matched_skills'] for skill in sublist]
top_skills = [skill for skill, _ in Counter(all_skills_flat).most_common(N)]

df_postings['top_skills'] = df_postings['matched_skills'].apply(lambda skills: [s for s in skills if s in top_skills])
mlb = MultiLabelBinarizer(classes=top_skills)
Y = mlb.fit_transform(df_postings['top_skills'])

features = ['industry_name', 'title', 'state', 'location']
df_model = df_postings.dropna(subset=features)
X = df_model[features]
Y = Y[df_model.index] 

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_encoded = encoder.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)

base_model = LGBMClassifier(n_estimators=100, max_depth=8, random_state=42)
multi_model = MultiOutputClassifier(base_model)
multi_model.fit(X_train, Y_train)

Y_pred = multi_model.predict(X_test)
report = classification_report(Y_test, Y_pred, target_names=top_skills, zero_division=0, output_dict=True)
report_df = pd.DataFrame(report).transpose()

print("🎯 Multi-label Skill Classification Report:")
display(report_df.head(12))
```

## Ablation study of Model 2.

```python
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

feature_sets = {
    "All Features": ['industry_name', 'title', 'state', 'location'],
    "No industry_name": ['title', 'state', 'location'],
    "No title": ['industry_name', 'state', 'location'],
    "No state": ['industry_name', 'title', 'location'],
    "No location": ['industry_name', 'title', 'state']
}

ablation_results = []

for label, selected_features in feature_sets.items():
    # Subset data
    X = df_postings.dropna(subset=selected_features)[selected_features]
    Y_subset = mlb.transform(df_postings.loc[X.index, 'top_skills']) 
    
    encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    X_encoded = encoder.fit_transform(X)
    
    X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y_subset, test_size=0.2, random_state=42)

    base_model = LGBMClassifier(n_estimators=100, max_depth=8, random_state=42)
    multi_model = MultiOutputClassifier(base_model)
    multi_model.fit(X_train, Y_train)
    
    Y_pred = multi_model.predict(X_test)
    f1_micro = f1_score(Y_test, Y_pred, average='micro', zero_division=0)
    f1_macro = f1_score(Y_test, Y_pred, average='macro', zero_division=0)
    
    ablation_results.append({
        "Feature Set": label,
        "F1 Micro": round(f1_micro, 4),
        "F1 Macro": round(f1_macro, 4)
    })

ablation_df = pd.DataFrame(ablation_results)
display(ablation_df)
```

## Model 3: Salary Prediction, setup.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_absolute_error, r2_score, root_mean_squared_error
from sklearn.preprocessing import OrdinalEncoder
from lightgbm import LGBMRegressor

df = pd.read_csv("refined_postings.csv")

required_cols = ['normalized_salary', 'industry_name', 'title', 'location', 'state', 'formatted_work_type', 'employee_count']
df_model = df.dropna(subset=required_cols)

Q1 = df_model['normalized_salary'].quantile(0.25)
Q3 = df_model['normalized_salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_model = df_model[(df_model['normalized_salary'] >= lower_bound) & (df_model['normalized_salary'] <= upper_bound)]

df_model['log_salary'] = np.log1p(df_model['normalized_salary'])

features = ['industry_name', 'title', 'location', 'state', 'formatted_work_type', 'employee_count']
target = 'log_salary'
X = df_model[features].copy()
y = df_model[target]

cat_features = ['industry_name', 'title', 'location', 'state', 'formatted_work_type']
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_encoded = X.copy()
X_encoded[cat_features] = encoder.fit_transform(X_encoded[cat_features])

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = LGBMRegressor(random_state=42)
param_grid = {
    'n_estimators': [100, 150],
    'max_depth': [6, 10],
    'learning_rate': [0.1, 0.05]
}

grid = GridSearchCV(model, param_grid, cv=3, scoring='neg_root_mean_squared_error', n_jobs=-1, verbose=1)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
y_pred_log = best_model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_true = np.expm1(y_test)

rmse = root_mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print("✅ Best Hyperparameters:", grid.best_params_)
print("📉 RMSE:", round(rmse, 2))
print("📉 MAE:", round(mae, 2))
print("📈 R² Score:", round(r2, 4))
```

## Ablation study of Model 3.

```python
print("\n🔬 Starting Ablation Study...\n")

feature_groups = {
    "All Features": ['industry_name', 'title', 'location', 'state', 'formatted_work_type', 'employee_count'],
    "No Title": ['industry_name', 'location', 'state', 'formatted_work_type', 'employee_count'],
    "No Industry": ['title', 'location', 'state', 'formatted_work_type', 'employee_count'],
    "No Location": ['industry_name', 'title', 'state', 'formatted_work_type', 'employee_count'],
    "Only Structured": ['industry_name', 'formatted_work_type', 'employee_count'],
    "Only Unstructured": ['title', 'location', 'state']
}

ablation_results = []

for name, feats in feature_groups.items():
    X_subset = df_model[feats].copy()

    subset_cats = [col for col in cat_features if col in feats]
    X_subset[subset_cats] = encoder.fit_transform(X_subset[subset_cats])

    X_train, X_test, y_train, y_test = train_test_split(X_subset, y, test_size=0.2, random_state=42)

    model = LGBMRegressor(**grid.best_params_, random_state=42)
    model.fit(X_train, y_train)

    y_pred_log = model.predict(X_test)
    y_pred = np.expm1(y_pred_log)
    y_true = np.expm1(y_test)

    rmse = root_mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    ablation_results.append({
        'Feature Set': name,
        'RMSE': round(rmse, 2),
        'MAE': round(mae, 2),
        'R² Score': round(r2, 4)
    })

ablation_df = pd.DataFrame(ablation_results)
display(ablation_df)
```
