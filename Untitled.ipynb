### Loading the CSV files

```python
import pandas as pd

postings = pd.read_csv("postings.csv")
job_skills = pd.read_csv("job_skills.csv")
job_industries = pd.read_csv("job_industries.csv")
benefits = pd.read_csv("benefits.csv")
salaries = pd.read_csv("salaries.csv")
companies = pd.read_csv("companies.csv")
employee_counts = pd.read_csv("employee_counts.csv")
company_specialities = pd.read_csv("company_specialities.csv")
company_industries = pd.read_csv("company_industries.csv")
industries = pd.read_csv("industries.csv")
skills = pd.read_csv("skills.csv")
```

### Merging job related data using 'job_id'

```python
df = postings.merge(job_skills, on="job_id", how="left") \
             .merge(job_industries, on="job_id", how="left") \
             .merge(benefits, on="job_id", how="left") \
             .merge(salaries, on="job_id", how="left")
```

### Merging data using 'company_id'

```python
df1 = df.merge(companies, on="company_id", how="left") \
        .merge(company_industries, on="company_id", how="left") \
        .merge(employee_counts, on="company_id", how="left")
```

### Merging more data using 'industr_id' and 'skill_abr'

```python
df2 = df1.merge(industries, on="industry_id", how="left") \
       .merge(skills, on="skill_abr", how="left")
```

### Formatting the 'expiry' column into the datetime format. Making sure to retain the original Unix timestamp but setting the date to today.

```python
from datetime import datetime

df2["expiry"] = pd.to_datetime(df2["expiry"])

print(df2["expiry"])
```

```python
today = datetime.now()

df2["expiry"] = df2["expiry"].apply(lambda x: x.replace(year=today.year, month=today.month, day=today.day))
```

### Assessing the columns

```python
list(df2.columns)
```

### Discarding unnecessary columns

```python
df3 = df2.drop(columns=["job_posting_url", "application_url", "application_type", "posting_domain", "sponsored", "fips", "skill_abr", "inferred","url", "follower_count"], errors="ignore")
```

### There are some duplicate values differentiated by 'x' and 'y'. The 'y' duplicate columns are used to fill in any missing values in the 'x' columns.

```python
for col in ['max_salary', 'min_salary', 'med_salary', 'description', 'pay_period', 'currency', 'compensation_type', 'zip_code']:
    x_col = f"{col}_x"
    y_col = f"{col}_y"
    
    df3[x_col] = df3[x_col].fillna(df3[y_col])
    
    df3.drop(columns=[y_col], inplace=True)
```

### Assessing the columns of the new processed dataset

```python
list(df3.columns)
```

### Calculating the estimated median salary only where the 'med_salary' is missing

```python
df3['med_salary_x'] = df3['med_salary_x'].fillna(
    df3[['min_salary_x', 'max_salary_x']].mean(axis=1)
)
```

### Missing values in the 'industry_name' column are filled in with the values from the 'industry' column

```python
df3['industry_name'] = df3['industry_name'].fillna(df3['industry'])
```

### More unnecessary columns are discarded

```python
df4 = df3.drop(columns=["max_salary_x", "min_salary_x", "company_id", "industry_id", "salary_id", "work_type", "views", "applies", "type", "name", "skill_name", "listed_time", "company_size", "time_recorded", "industry"], errors="ignore")
```

### Missing rows in relations to company name, job title, region, and salary has been discarded

```python
df5 = df4.dropna(subset=['company_name', 'title', 'location', 'normalized_salary'])
```

### Duplicate rows were also discarded, which is done with the help of 'job_id'.

```python
df6 = df5.drop_duplicates(subset='job_id', keep='first')
```

### Formatting the 'original_listed_time' column into the datetime format. Making sure to retain the original Unix timestamp but setting the date to today.

```python
df6["original_listed_time"] = pd.to_datetime(df6["original_listed_time"], errors='coerce')

print(df6["original_listed_time"])
```

### Exported the processed dataframe into a csv file to manually check the data

```python
df6.to_csv("refined_postings.csv", index=False)
```

### Below are some plots of bar charts and heatmaps for initial analysis

```python
import matplotlib.pyplot as plt

df6['location'].value_counts().head(20).sort_values().plot(kind='barh', figsize=(8, 6))
plt.grid(True, which='both', axis='both', linestyle='--', alpha=0.7)
plt.title("Top 20 Locations by Job Postings")
plt.xlabel("Number of Postings")
plt.tight_layout()
plt.savefig("job_postings_by_location.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
df6['company_name'].value_counts().head(20).sort_values().plot(kind='barh', figsize=(8, 6), color='seagreen')
plt.grid(True, which='both', axis='both', linestyle='--', alpha=0.7)
plt.title("Top 20 Companies by Job Postings")
plt.xlabel("Number of Postings")
plt.tight_layout()
plt.savefig("job_postings_by_company.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
df6['formatted_experience_level'].value_counts().plot(kind='bar', figsize=(8, 6), color='firebrick')
plt.grid(True, which='both', axis='both', linestyle='--', alpha=0.7)
plt.title("Job Experience Level Distribution")
plt.xlabel("Number of Postings")
plt.tight_layout()
plt.savefig("job_postings_by_experience_level.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import seaborn as sns

top_companies = df6['company_name'].value_counts().head(20).index
top_titles = df6['title'].value_counts().head(10).index
filtered_df6 = df6[df6['company_name'].isin(top_companies) & df6['title'].isin(top_titles)]
heatmap_data1 = filtered_df6.groupby(['company_name', 'title']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data1, annot=True, fmt='d', cmap='Purples')
plt.title("Heatmap: Top 20 Companies and Their Top 10 Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Company Name")
plt.tight_layout()
plt.savefig("companies_vs_job_titles.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
top_companies = df6['company_name'].value_counts().head(20).index
experience_level = df6['formatted_experience_level'].value_counts().index
filtered_df6 = df6[df6['company_name'].isin(top_companies) & df6['formatted_experience_level'].isin(experience_level)]
heatmap_data1 = filtered_df6.groupby(['company_name', 'formatted_experience_level']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data1, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Heatmap: Top 20 Companies and Their Expected Job Experience Levels")
plt.xlabel("Job Experience level")
plt.ylabel("Company Name")
plt.tight_layout()
plt.savefig("companies_vs_job_experience_level.png", dpi=300, bbox_inches='tight')
plt.show()
```
